```python
#to find duplicate
import pandas as pd
a=pd.read_csv('data3.csv')
print(a.duplicated())
```

```python
#to remove duplicate ,use the drop_duplicates() method
import pandas as pd
a=pd.read_csv('data3.csv')
a.drop_duplicates(inplace=True)
print(a.to_string())
```

```python
#to remove duplicate ,use the drop_duplicates() method
import pandas as pd
a=pd.read_csv('data3.csv')
a.drop_duplicates(inplace=True,keep='first')
print(a.to_string())
```

```python
#to remove duplicate ,use the drop_duplicates() method
import pandas as pd
a=pd.read_csv('data3.csv')
a.drop_duplicates(inplace=True,keep='last')
print(a.to_string())
```

```python
#to remove duplicate ,use the drop_duplicates() method
import pandas as pd
a=pd.read_csv('data3.csv')
a.drop_duplicates(inplace=True,keep=False)
print(a.to_string())
```

```python
#removings rows which are empty
import pandas as pd
a=pd.read_csv('data2.csv')
b=a.dropna()
print(b.to_string())
#saving the clean file
b.to_csv('newdata2.csv',index=False)
```

```python
#changing the existing dataframe
import pandas as pd
a=pd.read_csv('data3.csv')
a.dropna(inplace=True)
print(a.to_string())
#saving the clean file
a.to_csv('newdata3.csv',index=False)
```

```python
#replacing emplty value
import pandas as pd
a=pd.read_csv('data3.csv')
a.fillna(42,inplace=True)
print(a)
```

```python
#replacing emplty value for ds_masks
import pandas as pd
a=pd.read_csv('data3.csv')
a["DS_Marks"].fillna(49,inplace=True)
print(a)
```

```python
#using mean
#replacing emplty value
import pandas as pd
a=pd.read_csv('data3.csv')
b=a["DS_Marks"].mean()
a["DS_Marks"].fillna(b,inplace=True)
print(a)
print(b)
```

```python
#calculate median
#replacing emplty value
import pandas as pd
a=pd.read_csv('data3.csv')
b=a["Python_Marks"].median()
a["Python_Marks"].fillna(b,inplace=True)
print(a)
print(b)
```

```python
#calculate mode
#replacing emplty value
import pandas as pd
a=pd.read_csv('data3.csv')
b=a["Python_Marks"].mode()[0]
a["Python_Marks"].fillna(b,inplace=True)
print(a)
print(b)
```

```python
#changing data in specific row
import pandas as pd
a=pd.read_csv('data3.csv')
a.loc[1,'Python_Marks']=99
print(a.to_string)
```

```python
#replace values greater than 100 by 95
import pandas as pd
a=pd.read_csv('data3.csv')
for x in a.index:
    if a.loc[x,"Python_Marks"]>90:
        a.loc[x,"Python_Marks"]=85
print(a.to_string)
```

```python
#replace rows
import pandas as pd
a=pd.read_csv('data3.csv')
for x in a.index:
    if a.loc[x,"Python_Marks"]>90:
        a.drop(x,inplace=True)
print(a.to_string)
```

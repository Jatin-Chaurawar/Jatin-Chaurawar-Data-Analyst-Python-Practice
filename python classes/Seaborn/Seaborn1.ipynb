```python
#load a built in dataset
a=sns.load_dataset('car_crashes')
print(a)
```

```python
sns.distplot(a['not_distracted'])
```

```python
sns.distplot(a['not_distracted'],kde=False,bins=30)
```

```python
sns.histplot(a['not_distracted'])
```

```python
sns.histplot(a['not_distracted'],bins=30)
```

```python
#jointplot copares two distribution and plot a scatter plot by default
sns.jointplot(x='speeding',y='alcohol',data=a,kind='reg')
```

```python
#jointplot copares two distribution and plot a scatter plot by default
sns.jointplot(x='speeding',y='alcohol',data=a )
```

```python
sns.kdeplot(a['alcohol'])
```

```python
pair plot plot 
sns.pairplot(a)
```

```python
#rug plot
sns.rugplot(a['alcohol'])
```

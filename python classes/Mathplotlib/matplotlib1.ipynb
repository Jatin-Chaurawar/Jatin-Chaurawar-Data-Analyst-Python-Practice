```python
#drawing a line in a diagram from position(0,0) to position (6,50) 
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,50])
plt.plot(x,y)
plt.show()
```

```python
#adding title to the plot
import matplotlib.pyplot as plt
x=[1,4,3,7]
y=[3,8,1,10]
plt.plot(x,y,'*--g')
#adding title to the graph
plt.title("My First Graph",fontsize=20,color="green")
plt.show()
```

```python
#adding lablee to the plot
import matplotlib.pyplot as plt
x=[1,4,3,7]
y=[3,8,1,10]
plt.plot(x,y,'*-g')
#adding title to the graph
plt.title("My First Graph")
#adding lable on y axis
plt.ylabel('Y-Axis')
#adding lable on x axis
plt.xlabel('X-Axis')
plt.show()
```

```python
#using custom title and lable 
x=[80,85,90,95,100,105,110,115,120,125]
y=[240,250,260,270,280,290,300,310,320,330]
plt.title("My Graoh",loc='left',fontsize=10,color='r')
plt.xlabel("Hindi Marks",fontsize=10,color='k')
plt.ylabel("English Marks",fontsize=10,color='k')
plt.plot(x,y)
plt.show()
```

```python
#program to set marker size
import matplotlib.pyplot as plt
y=[2,5,1,17]
plt.plot(y,marker='o',ms=20)
plt.show()
```

```python
#program to set marker size
import matplotlib.pyplot as plt
x=[1,9,7,8]
y=[2,5,1,17]
plt.plot(x,y,marker='*',ms=20,mec='r',mfc='k')
plt.show()
```

```python
#dotted lines
#program to set marker size
import matplotlib.pyplot as plt
x=[1,9,7,8]
y=[2,5,1,17]
plt.plot(x,y,linestyle='dotted')
plt.show()
```

```python
#changing line color and width
import matplotlib.pyplot as plt
x=[4,19,12,30]
y=[3,8,1,10]
plt.plot(x,y,lw='10.5',c='r')
plt.show()
```

```python
# To draw two lines
import matplotlib.pyplot as plt
x=[4,19,12,30]
y=[3,8,1,10]
plt.plot(x,c='r',ls='dotted')
plt.plot(y,c='k',ls='dotted')
plt.show()
```

```python
#using custom title and lable 
x=[80,85,90,95,100,105,110,115,120,125]
y=[240,250,260,270,280,290,300,310,320,330]
plt.title("My Graoh",loc='left',fontsize=10,color='r')
plt.xlabel("Hindi Marks",fontsize=10,color='k')
plt.ylabel("English Marks",fontsize=10,color='k')
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#using custom title and lable 
x=[80,85,90,95,100,105,110,115,120,125]
y=[240,250,260,270,280,290,300,310,320,330]
plt.title("My Graoh",loc='left',fontsize=10,color='r')
plt.xlabel("Hindi Marks",fontsize=10,color='k')
plt.ylabel("English Marks",fontsize=10,color='k')
plt.plot(x,y)
plt.grid(axis='y',c='g',lw=1,ls='dashed')
plt.show()
```

```python
import numpy as np 
df= pd.read_csv("demo1.csv")
# extract demo column

x= df['Student_Id'].values
y1= df['DA_Marks'].values
y2= df['DS_Marks'].values
y3= df['Python_Marks'].values
y4= df['SQL_Marks'].values

# create subplot with diffrent axis

fig,axes= plt.subplots(2,2,figsize=(10,8))
# plot on the first subplot

axes[0,0].plot(x,y1,label='DA_Marks')
axes[0,0].set_title('DA_Marks')
axes[0,0].set_xlabel('Student_Id')
axes[0,0].set_ylabel('Student_Marks')
axes[0,0].legend()

axes[0,1].plot(x,y2,color='r',label='DS_Marks')
axes[0,1].set_title('DS_Marks')
axes[0,1].set_xlabel('Student_Id')
axes[0,1].set_ylabel('Student_Marks')
axes[0,1].legend()

axes[1,0].plot(x,y3,color='g',label='Python_Marks')
axes[1,0].set_title('Python_Marks')
axes[1,0].set_xlabel('Student_Id')
axes[1,0].set_ylabel('Student_Marks')
axes[1,0].legend()

axes[1,1].plot(x,y4,color='k',linestyle='--',label='SQL_Marks')
axes[1,1].set_title('SQL_Marks')
axes[1,1].set_xlabel('Student_Id')
axes[1,1].set_ylabel('Student_Marks')
axes[1,1].legend()

plt.tight_layout()
plt.show()
```

```python
df= pd.read_csv("demo1.csv")
a=df.sort_values(by='Python_Marks',ascending = False)
top_students=a.head(3)
plt.figure(figsize=(10,6))
plt.bar(top_students['Student_Id'],top_students['Python_Marks'],color='skyblue')
plt.xlabel("Student ID")
plt.ylabel("Python Marks")
plt.title("Top 5 Student based on Python Marks")
plt.grid(True,axis='y')
plt.tight_layout()
plt.show()
```

```python
df= pd.read_csv("demo1.csv")
#calculate total marks for each student
df['Total_Marks']=df[['Python_Marks','DS_Marks','DA_Marks','SQL_Marks']].sum(axis=1)
plt.figure(figsize=(10,6))
plt.bar(df['Student_Id'],df['Total_Marks'],color='skyblue')
plt.xlabel("Student ID")
plt.ylabel("Total Marks")
plt.title("Total Marks for Each Student")
plt.tight_layout()
plt.show()
```

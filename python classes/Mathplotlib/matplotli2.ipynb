```python
#to draw 6 plot in one figuure
import matplotlib.pyplot as plt
#plot1:
x=[1,2,3,4]
y=[3,8,1,10]
plt.subplot(2,3,1)
plt.title("Graph 1")
plt.plot(x,y)
#plot2:
x=[4,5,6,7]
y=[10,20,30,40]
plt.subplot(2,3,2)
plt.title("Graph 2")
plt.plot(x,y)
#plot3:
x=[8,9,10,11]
y=[4,10,31,34]
plt.subplot(2,3,3)
plt.title("Graph 3")
plt.plot(x,y)
#plot4:
x=[12,13,14,15]
y=[45,4,15,28]
plt.subplot(2,3,4)
plt.title("Graph 4")
plt.plot(x,y)
#plot5:
x=[16,17,18,19]
y=[56,89,13,19]
plt.subplot(2,3,5)
plt.title("Graph 5")
plt.plot(x,y)
#plot6:
x=[20,21,22,23]
y=[40,67,12,19]
plt.subplot(2,3,6)
plt.title("Graph 6")
plt.plot(x,y)
plt.suptitle('My Graph for 6 plots')
plt.show()
```

```python
#Assigning specific color to each scatter point
import matplotlib.pyplot as plt
#data foe 1st scatter plot
x=[5,7,8,7,2,17,2]
y=[99,103,87,94,78,77,85]
colors=['red','blue','black','gray','purple','cyan','maroon']
plt.scatter(x,y,c=colors)
plt.show()
```

```python
#using colormap
import matplotlib.pyplot as plt
#data foe 1st scatter plot
x=[5,7,8,7,2,17,2]
y=[99,103,87,94,78,77,85]
colors=[0,10,20,30,40,45,50,]
plt.scatter(x,y,c=colors,cmap='viridis')
plt.show()
```

```python
#to show colormap colormap in the figure 
import matplotlib.pyplot as plt
#data foe 1st scatter plot
x=[5,7,8,7,2,17,2]
y=[99,103,87,94,78,77,85]
colors=[0,10,20,30,40,45,50,]
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#to manage size of scatter graph dots
import matplotlib.pyplot as plt
#data foe 1st scatter plot
x=[5,7,8,7,2,17,2]
y=[99,103,87,94,78,77,85]
colors=[0,10,20,30,40,45,50]
sizes=[105,45,60,30,58,89,50]
#use values bettween 0.1 to 1
plt.scatter(x,y,s=sizes,alpha=0.5,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#using random with numpy
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(50,size=(50))
y=np.random.randint(50,size=(50))
colors=np.random.randint(50,size=(50))
sizes=10 * np.random.randint(50,size=(50))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#creating bargraph
import matplotlib.pyplot as plt
x=["A","B","C","D"]
y=[13,18,11,35]
plt.bar(x,y)
plt.show()
```

```python
#creating a bar graph with all details
import matplotlib.pyplot as plt
categories=['Category A','Category B','Category C','Category D',]
values=[25,30,20,35]
plt.bar(categories,values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title('My Bar Graph')
plt.show()
```

```python
#basic histogram
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=1000,color='skyblue',edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title('Basic Histogram')
plt.show()
```
